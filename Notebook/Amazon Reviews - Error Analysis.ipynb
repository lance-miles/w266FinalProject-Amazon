```python
import pandas as pd
import numpy as np
pd.options.display.max_colwidth = 2000
```

```python
zhang_yelp_reviews = pd.read_pickle('../../yelp_zhang_test_predictions (1).pkl')
yelp_reviews = pd.read_pickle('../../yelp_test_predictions (1).pkl')
twitter_sent140 = pd.read_pickle('../../twitter_sent140_test_predictions (1).pkl')
imdb_reviews = pd.read_pickle('../../imdb_test_predictions (1).pkl')
amazon_zhang = pd.read_pickle('../../amazon_zhang_test_predictions (1).pkl')
sst = pd.read_pickle('../../sst_test_predictions (1).pkl')
```

```python
zhang_yelp_reviews['Dataset'] = 'Zhang Yelp Reviews'
yelp_reviews['Dataset'] = 'Yelp Reviews'
twitter_sent140['Dataset'] = 'Twitter'
imdb_reviews['Dataset'] = 'IMDB Reviews'
amazon_zhang['Dataset'] = 'Zhang Amazon'
sst['Dataset'] = 'SST'
```

```python
sst.head()
```

```python
sentiment_df = pd.concat([zhang_yelp_reviews, yelp_reviews, twitter_sent140, imdb_reviews, amazon_zhang, sst], axis=0, sort=True)

sentiment_df = sentiment_df[['Dataset','sentiment','USE_Soft','CNN_TE','CNN_TE_USE','CNN_Glove','CNN_Glove_USE','CNN_w2V','CNN_w2V_USE','Individual_Sent','Average_Sent']]
sentiment_df.head()
```

```python
sentiment_df.to_csv('../../sentiment_df.csv')
```

```python
sentiment_df_long = pd.melt(sentiment_df, id_vars = ['Dataset','sentiment'], value_vars = ['USE_Soft','CNN_TE','CNN_TE_USE','CNN_Glove','CNN_Glove_USE','CNN_w2V','CNN_w2V_USE','Individual_Sent','Average_Sent'])
sentiment_df_long.head()
```

```python
sentiment_df_long.to_csv('../../sentiment_df_long.csv')
```
